# Normalización Alumnos y Claustro


```python
import pandas as pd
```

```python
claustro = pd.read_csv('data2/claustro.csv', index_col = False)
claustro
```

```python
claustro = claustro.drop(columns=["Vertical", "Campus", "Fecha_promocion"])
```

```python
claustro
```

```python
ID_intake = [1, 1, 2, 3, 4, 5, 1, 4, 2, 1]
```

```python
claustro= claustro.rename(columns={"id_claustro" : "ID_intake"})
```

```python
claustro["ID_Intake"] = ID_intake
```

```python
claustro
```

```python
claustro["ID_Claustro"] = range(1, len(claustro) + 1)
```

```python
claustro
```

```python
claustro=claustro.reindex(columns=["ID_Claustro", "ID_Intake", "Nombre", "Rol"])
claustro.columns
```

```python
claustro.to_csv('data3/claustro.csv', index=False)
```

```python
alumnos_DS = pd.read_csv('data2/listado_Ds.csv')
alumnos_DS.head()
```

```python
alumnos_FS = pd.read_csv('data2/listado_FSb.csv')
alumnos_FS.head()
```

```python
alumnos = pd.concat([alumnos_DS, alumnos_FS])
alumnos.head()
```

```python
alumnos["ID_Estudiante"] = range(1, len(alumnos) + 1)
```

```python
alumnos.to_csv('data2/alumnos.csv', index=False)
```

```python
alumnos = alumnos.iloc[:, :3]
```

```python
alumnos.head()
```
